```python
#code to find prima factors of any number
def prime_factors(n):
    factors = []
    divisor = 2

    while n > 1:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 1

    return factors

# Example usage:
number_to_factorize = 213335143247
factors = prime_factors(number_to_factorize)

print(f"Prime factors of {number_to_factorize}: {factors}")
```
